#### Subplots and Legends

```python
days = [1, 2, 3, 4, 5]
original_price = [10 ,20, 25, 30, 40]
predicted_price = [8, 21, 25, 28, 38]
```

```python
plt.plot(days, original_price, "b-o", label="Original Price")
plt.plot(days, predicted_price, "r-o", label="Predicted Price")
plt.legend()
plt.title("Plot with legends")
plt.xlabel("Weekdays")
plt.ylabel("Price")
```

#### Subplots

```python
# First subplot - Original Price
plt.subplot(3, 1, 1)
plt.plot(days, original_price, "b-o", label="Original Price")
plt.title("Original prices over time")
plt.title("Plot with legends")
plt.xlabel("Weekdays")
plt.ylabel("Price")
plt.legend()

# Second subplot - Predicted Price
plt.subplot(3, 1, 2)
plt.plot(days, predicted_price, "r-o", label="Predicted Price")
plt.title("Predicted prices over time")
plt.title("Plot with legends")
plt.xlabel("Weekdays")
plt.ylabel("Price")
plt.legend()

# Third subplot - Difference between Original and Predicted Prices
difference_price = [0 - p for o, p in zip(original_price, predicted_price)]
plt.subplot(3, 1, 3)
plt.bar(days, difference_price, color='purple', label="Difference in Prices")
plt.title("Differences in time")
plt.xlabel("Days")
plt.ylabel("Price Difference")
plt.legend()

plt.tight_layout()
```

#### Saving the Figure

```python
plt.plot(days, original_price, "b-o", label="Original Price")
plt.plot(days, predicted_price, "r-o", label="Predicted Price")
plt.legend()
plt.title("Plot with legends")
plt.xlabel("Weekdays")
plt.ylabel("Price")
plt.savefig("example_prediction.png", dpi=300)
```
